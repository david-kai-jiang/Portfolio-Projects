```python
# import pandasql library to use SQL 
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
# 1. Does a higher number of participating athletes correlate with  a higher medal count?
num_athletes_country = pysqldf('''SELECT regions.region, COUNT(*) AS `cnt`
                                FROM athlete_events
                                LEFT JOIN noc_regions AS regions
                                ON athlete_events.NOC = regions.NOC
                                GROUP BY regions.region
                                ORDER BY `cnt` DESC''')
num_medals_country = pysqldf('''SELECT regions.region, COUNT(*) AS `cnt`
                                FROM athlete_events
                                LEFT JOIN noc_regions AS regions
                                ON athlete_events.NOC = regions.NOC
                                WHERE athlete_events.Medal <> "None"
                                GROUP BY regions.region
                                ORDER BY `cnt` DESC''')
```

```python
athletes_medals_ratio = pysqldf('''SELECT athlete.region, athlete.cnt AS athlete_cnt, medal.cnt as medal_cnt, 
                                    (athlete.cnt * 1.0 / medal.cnt) AS ratio
                                FROM num_athletes_country as athlete
                                LEFT JOIN num_medals_country as medal
                                ON athlete.region = medal.region
                                ORDER BY athlete.cnt DESC
                                LIMIT 20''')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(20,10))
plt.plot(athletes_medals_ratio.region, athletes_medals_ratio.athlete_cnt, label = "Athletes Count")
plt.plot(athletes_medals_ratio.region, athletes_medals_ratio.medal_cnt, label = "Medals Count")
plt.title("Count of Athletes and Medals by Country", fontname = 'DejaVu Sans', fontsize = 14,
    fontstyle = 'italic', fontweight = 'bold')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# 4. Which country has the best number of medals to number of athletes ratio?
athletes_medals_ratio = pysqldf('''SELECT athlete.region, athlete.cnt AS athlete_cnt, medal.cnt as medal_cnt, 
                                    (medal.cnt * 1.0 / athlete.cnt) AS ratio
                                FROM num_athletes_country as athlete
                                LEFT JOIN num_medals_country as medal
                                ON athlete.region = medal.region
                                WHERE medal_cnt IS NOT null
                                ORDER BY ratio DESC
                                LIMIT 20''')
```

```python
athletes_medals_ratio.plot.bar(x='region', y='ratio', title = 'Number of Medals to Number of Athletes Ratio')
```

```python
# 5. Which sport gives the possibility to win the most medals?
sport_summer = pysqldf('''SELECT `Sport`, COUNT(*) AS `cnt`
                        FROM athlete_events
                        WHERE `Medal` IS NOT null AND `Season` = "Summer"
                        GROUP BY `Sport`
                        ORDER BY `cnt` DESC
                        LIMIT 25''')
sport_winter = pysqldf('''SELECT `Sport`, COUNT(*) AS `cnt`
                        FROM athlete_events
                        WHERE `Medal` IS NOT null AND `Season` = "Winter"
                        GROUP BY `Sport`
                        ORDER BY `cnt` DESC
                        LIMIT 25''')
```

```python
sport_summer.plot.bar(x='Sport', y='cnt', title = 'Summer Sports with the Most Medals')
```

```python
sport_winter.plot.bar(x='Sport', y='cnt', title = 'Winter Sports with the Most Medals')
```

```python
# import library to generate wordcloud
from wordcloud import WordCloud
```

```python
# generate wordcloud for summer sports with most medals
from collections import Counter

sport_summer_word = pysqldf('''SELECT `Sport` FROM athlete_events WHERE `Medal` IS NOT null AND `Season` = "Summer"''')

wordDict = Counter(sport_summer_word.Sport)
wordcloud = WordCloud().generate_from_frequencies(wordDict)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

```python
# generate wordcloud for winter sports with the most medals
sport_summer_word = pysqldf('''SELECT `Sport` FROM athlete_events WHERE `Medal` IS NOT null AND `Season` = "Winter"''')

wordDict = Counter(sport_summer_word.Sport)
wordcloud = WordCloud().generate_from_frequencies(wordDict)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
